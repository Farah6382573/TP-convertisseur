```python
# Code de la fonction à tester : pour calculer la somme des diviseurs inférieurs à n
def somme_quoi_inferieurs(n):
    somme = 0
    for i in range(1,n/2,1):
        if n % i == 0:
            somme += i
    return somme
```

### Travail demandé :
Complétez les **tests unitaires** pour cette fonction.

- Tester un nombre premier  
- Tester un nombre parfait (ex: 6)  
- Tester un cas limite (n = 1)  
- Tester un type invalide (si pertinent)  

Utilisez `pytest` ou `unittest`.


````python
import unittest
def somme_quoi_inferieurs(n):
    somme = 0
    for i in range(1, n//2 + 1):   # division entière corrigée
        if n % i == 0:
            somme += i
    return somme
# Classe de tests unitaires hérite de unittest.TestCase
class TestSommeDiviseurs(unittest.TestCase):
# Test d’un nombre premier  7 est premier → seul diviseur strict = 1
    def test_nombre_premier(self):
        self.assertEqual(somme_quoi_inferieurs(7), 1)
# Test d’un nombre parfait 6 → 1 + 2 + 3 = 6
    def test_nombre_parfait(self):
        self.assertEqual(somme_quoi_inferieurs(6), 6)
 # Test du cas limite 1 n’a aucun diviseur strict → somme = 0
    def test_cas_limite(self):
        self.assertEqual(somme_quoi_inferieurs(1), 0)
 # Test d’un type invalide :la fonction doit lever une erreur TypeError
    def test_type_invalide(self):
        with self.assertRaises(TypeError):
            somme_quoi_inferieurs("abc")


# Exécution (compatible Jupyter)
unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestSommeDiviseurs))
````

## Exercice 2 (/5 pts)

Soit le bloc de code suivant :

```python
# Créez une fonction simple à tester utiliser la classe TestCase.
def addition(a,b):
    return a+b-1

# Test unitaire sans la classe
def test_addition():
    result = addition(2,3)
```

### Travail demandé :
Ajoutez les instructions de test unitaire au code en utilisant la classe `TestCase`.

- Créez une classe héritant de `unittest.TestCase`
- Écrivez au moins 3 tests
- Montrez qu’un test échoue à cause de l’erreur logique dans `addition`
- Proposez la correction de la fonction


````python
import unittest
# Fonction  erronée
def addition(a, b):
    return a + b - 1


class TestAddition(unittest.TestCase):

    def test_addition_positive(self):
        # Ce test va échouer à cause du -1
        self.assertEqual(addition(2, 3), 5)

    def test_addition_zero(self):
        # 0 + 4 devrait donner 4 → échoue aussi
        self.assertEqual(addition(0, 4), 4)

    def test_addition_negative(self):
        # -2 + 5 = 3 → échoue également
        self.assertEqual(addition(-2, 5), 3)
# Exécution des tests
unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestAddition))

# Correction de la fonction
def addition_corrigee(a, b):
    return a + b

class TestAdditionCorrigee(unittest.TestCase):

    def test_addition_positive_corrigee(self):
        self.assertEqual(addition_corrigee(2, 3), 5)

    def test_addition_zero_corrigee(self):
        self.assertEqual(addition_corrigee(0, 4), 4)

    def test_addition_negative_corrigee(self):
        self.assertEqual(addition_corrigee(-2, 5), 3)

# Exécution des tests
unittest.TextTestRunner().run(unittest.TestLoader().loadTestsFromTestCase(TestAdditionCorrigee))
````
